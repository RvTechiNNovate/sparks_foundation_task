```python
# Here importing the required library for model
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read csv file
student_data = pd.read_csv('student_scores.csv') 
```

```python
student_data.head()
```

```python
# shape or dimention of student_data
student_data.shape
```

```python
student_data.plot(x='Hours', y='Scores', style='o') 
plt.title('Hours Studied v/s Percentage Score')
plt.xlabel('Hours student studied')
plt.ylabel("% of student's score") 
plt.show()
```

```python
# divide into featur and target
X = student_data.iloc[:,:-1].values
y = student_data.iloc[:, 1].values
```

```python
# split into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test =  train_test_split(X,y,random_state=0)
```

```python
# import linear regression model to train
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
# prediction
prediction=lr.predict(X_train)
```

```python
# training linear line visualization
plt.plot(X_train,prediction)
plt.title('Hours Studied v/s Percentage Score')
plt.xlabel('Hours student studied')
plt.ylabel("student's score (in %)") 
plt.show()
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
# Actual and prediction values in DataFrame
df_actual_vs_pred = pd.DataFrame({'Actual Value': y_test, 'Predicted value': y_pred})
df_actual_vs_pred.head()
```

```python
df_actual_vs_pred.plot(kind="bar", figsize=(10,8))
plt.xlabel("Hours student studied")
plt.ylabel("student's score (%)") 
plt.title("Hours Studied v/s % Score")
plt.show()
```

```python
train_score=lr.score(X_train,y_train)
test_score=lr.score(X_test,y_test)
print("Train Score: ",train_score)
print("Test Score: ",test_score)
```

```python
hrs = [7] 
ans = lr.predict([hrs])
print("Hours study = ",hrs)
print("Predicted Score = ",ans[0])

```

#### if a student study for 7 hour then he can score 71.52 % aprox by training this dataset
